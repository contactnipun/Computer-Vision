#### 2. Curating ccordinates/annotations

```python
x_cord=[]
y_cord=[]
height=[]
width=[]
face_detected=[]
image_name=[]
```

```python
# Open cv version 2

import cv2 

for i in range (len(dir_list)):
    print("Execution loop=",i)
    # Create a cascade classfier object
    face_cascade=cv2.CascadeClassifier("FF_Default.xml")   # frontal face 
    data_path = os.path.join(img_dir,dir_list[i])
    img=cv2.imread(data_path)
    faces=face_cascade.detectMultiScale(img,scaleFactor=1.05,minNeighbors=5)

    j=0 

    # Find coordinates of the face
    for x,y,w,h in faces:
        img=cv2.rectangle(img,(x,y),(x+w,y+h),(255, 0, 0),2)
        j=j+1
    resized=cv2.resize(img,(int(img.shape[1]/2),int(img.shape[0])))
    
    length=len(faces)
    if length==0:
        a=0
        b=0
        c=0
        d=0
        j=0
        name=0
        
    else:
        a=faces[0,0]
        b=faces[0,1]
        c=faces[0,2]
        d=faces[0,3]
        name=dir_list[i]
        
    
    x_cord.append(a)
    y_cord.append(b)
    width.append(c)
    height.append(d)
    face_detected.append(j)
    image_name.append(name)
```

```python
import pandas as pd

DB = pd.DataFrame(x_cord, columns = ['x'])

DB['y']=y_cord
DB['w']=width
DB['h']=height
DB['Total_Faces']=face_detected
DB['Image_Name']=image_name

DB.head()
```

#### 3. Exporting the coordinates/annotations to excel, csv or json file

```python
DB.to_excel("Bounding boxes.xlsx")
DB.to_csv("Bounding boxes.csv")
DB.to_json("Bounding boxes.json")
```
